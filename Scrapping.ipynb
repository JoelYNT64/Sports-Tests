```python
AGK_url = "https://stathead.com/fbref/team-match-finder.cgi?request=1&year_min=2024-2025&is_qualifier=0&comp_gender=m&match=team_game&force_year_min=1&match_status=1&comp_type=b5&timeframe=seasons&order_by=gk_passes&year_max=2024-2025&offset="
url= AGK_url+ "0"
driver.get(url)

time.sleep(5)  # Wait for the page to load



# Get page source after login
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
table = soup.find("table")

table
if table:
    head = table.find_all("thead")
    for row in head:
        cells = row.find_all("th")
        col_AGK = [cell.text.strip() for cell in cells]
        print(col_AGK)
```

```python
row_AGK_data= []
for page in range(0, 2800, 200):
    url= AGK_url + str(page)
    driver.get(url)

    time.sleep(5)  # Wait for the page to load

# Get page source after login
    html = driver.page_source
    time.sleep(3)
    soup = BeautifulSoup(html, "html.parser")
    table = soup.find("table")
    body= table.find_all("tbody")
    rows= body[0].find_all("tr")
    for row in rows:
        cells= row.find_all(["th", "td"])
        data= [cell.text.strip() for cell in cells]
        if data[0] not in ['', 'Rk']:
            row_AGK_data.append(data) 
```

```python
df_AGK= pd.DataFrame(row_AGK_data, columns= col_AGK)
df_AGK
```

```python
df_AGK.to_csv("../Data/Passes (GK).csv", index= False)
```
