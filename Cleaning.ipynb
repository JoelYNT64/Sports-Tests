```python
dfs= [df_AGK, df_ArW, df_AtD, df_Car, df_CrA, df_Crd, df_GcA, df_GKA, df_Gls, df_Misc, df_OPA, df_Pas, df_PkA, df_PkD, 
      df_PsT, df_Rec, df_ScA, df_ShA, df_Tch, df_Tkl, df_xG, df_xGK]#df_Tkl
common_cols = set(dfs[0].columns)
df_AGK.sort_values(by=["Team", "Opp", "Date"], inplace= True)
df_AGK.reset_index(drop=True, inplace= True) 
for df in dfs[1:]:
    df.sort_values(by=["Team", "Opp", "Date"], inplace=True)
    df.reset_index(drop=True, inplace= True)
    common_cols &= set(df.columns)
common_cols = list(common_cols)  # Convert to list
common_cols
```

```python
df_cols= df_ArW[common_cols]
for df in dfs:
    df.drop(columns=common_cols, inplace= True)
dfs.append(df_cols)
```

```python
df_Teams_Games= pd.concat([df for df in dfs], axis=1)
df_Teams_Games
```

```python
cols = common_cols + [col for col in df_Teams_Games.columns if col not in common_cols]
df1= df_Teams_Games[cols]
df1
```

```python
id= ["Team", "Opp", "Date", "Result", "Unnamed: 4", "GF", "GA", "GD"]
cols2= id+ [col for col in df1.columns if not(col in id)]
df2= df1[cols2].copy()
df2
```

```python
cols_to_delete= ["Match Report", "Rk"]
df2["Date"] = pd.to_datetime(df2["Date"], format="%Y-%m-%d")
df3 = df2.drop(columns= cols_to_delete)
df3
```

```python
df4 = df3.rename(columns={"Unnamed: 4": "H/A"})
df4
```

```python
df4["H/A"] = df4["H/A"].fillna("H").replace("@", "A")
df4
```

```python
df4["Result"] = df4["Result"].str[0]
df4
```

```python
df4.sort_values(by="Date", inplace=True)
df4
```

```python
df_league = {comp_value: df4[df4["Comp"] == comp_value] for comp_value in df4["Comp"].unique()}
```
